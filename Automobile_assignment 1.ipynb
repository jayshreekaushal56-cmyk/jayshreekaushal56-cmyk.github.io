```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Automobile_data[1].csv")
```

```python
df.head(3)
```

1. what is the average price and average mileage for convertible cars

```python
df.groupby(["body-style"])[["price","average-mileage"]].agg(["mean"])
```

```python
convertible_cars=df[df["body-style"]=="convertible"]
convertible_cars
```

```python
price= convertible_cars["price"].mean()
print(f"Average price for convertible cars : {price}")
```

```python
mileage=convertible_cars["average-mileage"].mean()
print(f"Average mileage for convertible cars : {mileage}")
```

2. Sort cars by company and price in decending order

```python
decen_ordr=df.sort_values(by=["company", "price"], ascending=[False, False])
print(f"Cars sorted by company and price : \n{decen_ordr}")
```

4. which company is producing most expensive car?

```python
mx=df.groupby(["company"])["price"].max()
print(f"Company producing most expensive car:\n{mx}")
```

```python
mx = df.groupby("company")["price"].max()
print(f"Company producing most expensive car:{mx.idxmax()}")
```

5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower

```python
length=df.nlargest(1,"length")[["company"]]
print(f"Cars on the basis of maximum length :\n {length}")
```

```python
wheel_base=df.nlargest(1,"wheel-base")[["company"]]
print(f"Cars on the basis of maximum wheel base :\n {wheel_base}")
```

```python
horsepower=df.nlargest(1,"horsepower")[["company"]]
print(f"Cars on the basis of maximum horsepower :\n {horsepower}")
```

```python
df.groupby(["company"])[["length","wheel-base","horsepower"]].max()
```

6. company wise average price of cars and order them in decending order


```python
df.groupby(["company"])[["price"]].mean().sort_values(by="price", ascending=False)
```

7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy

```python
df["decision"] = np.where((df["wheel-base"] < 100) & (df["length"] < 175),"dont buy","buy")
```

8. Select only cars whose decision is "buy"; find average price based on company of car

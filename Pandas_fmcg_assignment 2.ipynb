```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Raw_Sales_Data.csv")
```

```python
df.head(3)
```

```python
df.info()
```

1. Monthwise number of quantities sold

```python
df["Qty"]=df["Qty"].str.replace(",","").astype(int)
```

```python
months = df.groupby("Month",as_index=False)["Qty"].sum()
months
```

2. Category wise Qty sold


```python
Category=df.groupby("Category",as_index=False)["Qty"].sum()
Category
```

3. Sub-channel wise total sales


```python
df["Sales"]=df["Sales"].str.replace(",","").astype(int)
```

```python
sales = df.groupby("Sub-Channels", as_index=False)["Sales"].sum()
sales
```

4. Highest sale in which month


```python
Highest_sale=df.nlargest(1,"Sales")[["Month"]]
Highest_sale
```

```python
Highest_sale=df.groupby("Month")["Sales"].max()
Highest_sale
```

5. Highest selling product (based on sales) (packaging quantity doesn't matter. Ex. consider "GET SLIM
JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)

```python
Highest_selling_product=df.groupby("Product Name")["Sales"].max()
Highest_selling_product
```

6. List of different products company is offering (packaging quantity doesn't matter. Ex. consider "GET
SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)


```python
print(f"List of different products company is offering:\n{df["Product Name"].unique()}")
```
